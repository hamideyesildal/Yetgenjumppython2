random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
import time

class Kumanda:
    def __init__(self):
        self.tv_durumu = False
        self.ses = 0
        self.kanal_listesi = ["TRT", "CNN", "FOX", "NTV", "Bloomberg"]
        self.kanal = "TRT"
    
    def tv_ac(self):
        if self.tv_durumu:
            print("Televizyon zaten açık.")
        else:
            print("Televizyon açılıyor...")
            self.tv_durumu = True
    
    def tv_kapat(self):
        if self.tv_durumu:
            print("Televizyon kapanıyor...")
            self.tv_durumu = False
        else:
            print("Televizyon zaten kapalı.")
    
    def ses_ayarları(self):
        while True:
            cevap = input("Sesi açmak için '+', kısmak için '-', çıkmak için 'q' tuşuna basın:")
            if cevap == "+":
                if self.ses != 32:
                    self.ses += 1
                    print("Ses:", self.ses)
            elif cevap == "-":
                if self.ses != 0:
                    self.ses -= 1
                    print("Ses:", self.ses)
            else:
                print("Ses güncellendi:", self.ses)
                break
    
    def kanal_ekle(self, kanal_adi):
        self.kanal_listesi.append(kanal_adi)
        print("Yeni kanal eklendi:", kanal_adi)
    
    def kanal_sayisi(self):
        print("Toplam kanal sayısı:", len(self.kanal_listesi))
    
    def rastgele_kanal(self):
        rastgele_sayi = random.randint(0, len(self.kanal_listesi)-1)
        self.kanal = self.kanal_listesi[rastgele_sayi]
        print("Şu anki kanal:", self.kanal)
    
    def tv_bilgileri(self):
        print("Televizyon Durumu:", "Açık" if self.tv_durumu else "Kapalı")
        print("Ses Düzeyi:", self.ses)
        print("Şu anki Kanal:", self.kanal)

kumanda = Kumanda()

print("""
Televizyon Kumandası Uygulaması

1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri

Çıkmak için q tuşuna basın.
""")

while True:
    islem = input("İşlem Seçiniz:")
    if islem == "q":
        print("Kumanda kapatılıyor...")
        break
    elif islem == "1":
        kumanda.tv_ac()
    elif islem == "2":
        kumanda.tv_kapat()
    elif islem == "3":
        kumanda.ses_ayarları()
    elif islem == "4":
        kanal_isimleri = input("Kanal isimlerini ',' ile ayırarak girin:")
       

```
