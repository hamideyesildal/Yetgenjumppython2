```python
class MenuItem:
    def __init__(self, name, price, calories):
        self.name = name
        self.price = price
        self.calories = calories

menu_items = []

try:
    menu_file = input("Enter menu file name: ")
    with open(menu_file, 'r') as f:
        for line in f:
            name, price, calories = line.strip().split(',')
            menu_item = MenuItem(name, float(price), int(calories))
            menu_items.append(menu_item)

except FileNotFoundError:
    print("Error: File not found")
    exit()

def find_menu_item(name):
    for menu_item in menu_items:
        if menu_item.name == name:
            return menu_item
    return None

while True:
    print("\nSelect an option:")
    print("1. Find price of a menu item")
    print("2. Find calories of a menu item")
    print("3. Quit")

    option = input("Enter option number: ")
    if option == '1':
        item_name = input("Enter menu item name: ")
        menu_item = find_menu_item(item_name)
        if menu_item:
            print(f"{item_name} costs {menu_item.price} dollars")
        else:
            print(f"{item_name} not found in menu")
    elif option == '2':
        item_name = input("Enter menu item name: ")
        menu_item = find_menu_item(item_name)
        if menu_item:
            print(f"{item_name} has {menu_item.calories} calories")
        else:
            print(f"{item_name} not found in menu")
    elif option == '3':
        print("Goodbye!")
        break
    else:
        print("Invalid option, please try again.")
```

```python
import csv

class Person:
    def __init__(self, name, age, gender, occupation):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupation = occupation

people = []

try:
    csv_file = input("Enter CSV file name: ")
    with open(csv_file, 'r') as f:
        reader = csv.reader(f)
        next(reader)  # skip header row
        for row in reader:
            name, age, gender, occupation = row
            person = Person(name, int(age), gender, occupation)
            people.append(person)

except FileNotFoundError:
    print("Error: File not found")
    exit()

def get_occupations():
    occupations = {}
    for person in people:
        if person.occupation in occupations:
            occupations[person.occupation] += 1
        else:
            occupations[person.occupation] = 1
    return occupations

def get_gender_count():
    male_count = 0
    female_count = 0
    for person in people:
        if person.gender == 'Male':
            male_count += 1
        elif person.gender == 'Female':
            female_count += 1
    return male_count, female_count

def get_age_range():
    min_age = float('inf')
    max_age = float('-inf')
    for person in people:
        if person.age < min_age:
            min_age = person.age
        if person.age > max_age:
            max_age = person.age
    return min_age, max_age

print("Occupation counts:")
occupations = get_occupations()
for occupation, count in occupations.items():
    print(f"{occupation}: {count}")

male_count, female_count = get_gender_count()
print(f"Male: {male_count}, Female: {female_count}")

min_age, max_age = get_age_range()
print(f"Youngest person: {min_age}, Oldest person: {max_age}")
```

```python
import pandas as pd


class VeriAnalizi:
    def __init__(self, mutluluk):
        self.dosya_adi = mutluluk
        self.veri = pd.read_excel(mutluluk)

    def satir_sayisi(self):
        return len(self.veri)

    def sutun_sayisi(self):
        return len(self.veri.columns)

    def sutun_adi(self):
        return list(self.veri.columns)

    def sutun_verileri(self, sutun):
        return self.veri[sutun].tolist()

    def ortalama(self, sutun):
        return self.veri[sutun].mean()

    def medyan(self, sutun):
        return self.veri[sutun].median()

    def mod(self, sutun):
        return self.veri[sutun].mode().tolist()[0]

    def varyans(self, sutun):
        return self.veri[sutun].var()

    def standart_sapma(self, sutun):
        return self.veri[sutun].std()

    def min(self, sutun):
        return self.veri[sutun].min()

    def max(self, sutun):
        return self.veri[sutun].max()

    def giris(self):
        while True:
            try:
                sutun = input("Lütfen bir sutun adi girin: ")
                if sutun not in self.sutun_adi():
                    raise ValueError("Geçersiz sutun adi!")
                break
            except ValueError as e:
                print(str(e))
        return sutun


if __name__ == "__main__":
    while True:
        try:
            dosya_adi = input("Lütfen bir dosya adi girin: ")
            veri_analizi = VeriAnalizi(dosya_adi)
            break
        except FileNotFoundError:
            print("Hata: Dosya bulunamadi! Lütfen gecerli bir dosya adi girin.")

    print("Veri Analizi")
    print("============")
    print(f"Toplam Satır Sayısı: {veri_analizi.satir_sayisi()}")
    print(f"Toplam Sütun Sayısı: {veri_analizi.sutun_sayisi()}")

    while True:
        sutun = veri_analizi.giris()
        print("Seçilen Sütun: ", sutun)
        print(f"Ortalama: {veri_analizi.ortalama(sutun)}")
        print(f"Medyan: {veri_analizi.medyan(sutun)}")
        print(f"Mod: {veri_analizi.mod(sutun)}")
        print(f"Varyans: {veri_analizi.varyans(sutun)}")
        print(f"Standart Sapma: {veri_analizi.standart_sapma(sutun)}")
        print(f"Minimum: {veri_analizi.min(sutun)}")
        print(f"Maksimum: {veri_analizi.max(sutun)}")

        cev
```

```python
class OgrenciNotlari:
    def __init__(self, ad, soyad):
        self.ad = ad
        self.soyad = soyad
        self.notlar = []

    def not_ekle(self, notu):
        self.notlar.append(notu)

    def not_ortalamasi(self):
        return sum(self.notlar) / len(self.notlar)


def not_girme():
    while True:
        try:
            ad = input("Öğrenci adını girin: ")
            soyad = input("Öğrenci soyadını girin: ")
            ogrenci = OgrenciNotlari(ad, soyad)
            not_sayisi = int(input("Kaç adet not gireceksiniz? "))
            for i in range(not_sayisi):
                notu = int(input(f"{i+1}. notu girin: "))
                ogrenci.not_ekle(notu)
            print(f"{ogrenci.ad} {ogrenci.soyad}'in not ortalaması: {ogrenci.not_ortalamasi()}")
            break
        except ValueError:
            print("Lütfen geçerli bir sayı girin!")


not_girme()
```

```python
import random

class Oyun:
    def __init__(self, isim):
        self.isim = isim
        self.puan = 0
    
    def baslat(self):
        print(f"{self.isim} adlı oyuna hoş geldiniz!")
        while True:
            sayi = random.randint(1, 10)
            tahmin = input("1 ile 10 arasında bir sayı tahmin edin: ")
            try:
                tahmin = int(tahmin)
            except ValueError:
                print("Geçerli bir sayı girin!")
                continue
            if tahmin == sayi:
                self.puan += 10
                print("Tebrikler! Doğru tahmin ettiniz.")
            else:
                print(f"Üzgünüm, doğru sayı {sayi} idi.")
            devam = input("Tekrar oynamak ister misiniz? (e/h): ")
            if devam.lower() != "e":
                print(f"{self.isim} oyunundan {self.puan} puan aldınız. İyi oyunlar!")
                break

def oyunu_baslat():
    isim = input("Oyuncu adınızı girin: ")
    oyun = Oyun(isim)
    oyun.baslat()

oyunu_baslat()
```

```python
import random

class SayiTahminOyunu:
    def __init__(self, tahmin_hakki):
        self.tahmin_hakki = tahmin_hakki
        self.sansli_sayi = random.randint(1, 100)
    
    def puan_hesapla(self, kalan_tahmin):
        return 100 - (self.tahmin_hakki - kalan_tahmin) * 20
    
    def oyunu_baslat(self):
        print(f"1 ile 100 arasında bir sayı tahmin et. Toplam {self.tahmin_hakki} tahmin hakkın var.")
        kalan_tahmin = self.tahmin_hakki
        while kalan_tahmin > 0:
            try:
                tahmin = int(input("Tahminin nedir? "))
            except ValueError:
                print("Lütfen bir sayı girin.")
                continue
            
            kalan_tahmin -= 1
            if tahmin == self.sansli_sayi:
                print(f"Tebrikler! {self.sansli_sayi} sayısını {self.puan_hesapla(kalan_tahmin)} puanla buldunuz.")
                break
            elif tahmin < self.sansli_sayi:
                print("Daha büyük bir sayı söyleyin.")
            else:
                print("Daha küçük bir sayı söyleyin.")
        else:
            print(f"Maalesef hakkınız bitti. Doğru cevap {self.sansli_sayi} idi.")

def sayi_tahmin_oyunu_baslat():
    try:
        tahmin_hakki = int(input("Kaç tahmin hakkı istersiniz? "))
    except ValueError:
        print("Lütfen bir sayı girin.")
        return
    
    sayi_tahmin_oyunu = SayiTahminOyunu(tahmin_hakki)
    sayi_tahmin_oyunu.oyunu_baslat()

sayi_tahmin_oyunu_baslat()
```
